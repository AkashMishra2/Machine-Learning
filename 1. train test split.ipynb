## train test split

```python
import pandas as pd
import seaborn as sns 
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
# taking only numerical features of dataset
df2 = df[["survived","pclass","age","sibsp","parch","fare"]]
```

```python
print(df2.isnull().sum())
df2.head()
```

```python
# need to clean dataset using median value 
df3 = df2.fillna(df2.median())
df3.isnull().sum()
```

```python
df3.head()
```

```python
# converting data in X,Y   --> X independent variable, Y is target variable or dependent variable 
X = df3.drop(["survived"],axis = 1)
y = df3["survived"]
```

```python
print("Shape of X: ",X.shape)
print("Shape of y: ",y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=51)
# without random_state our train and test data is shuffling 
# for not shuffling the data we must have to use random_size and give any number

print("Shape of X_train: ",X_train.shape)
print("Shape of X_test: ",X_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test: ",y_test.shape)
```
