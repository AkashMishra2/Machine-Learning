```python
# importing libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression  # for Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures  # for Polynomial Regression
from sklearn.metrics import mean_squared_error
```

```python
# loading dataset
df = pd.read_csv(r"D:\bangalore house price prediction OHE-data.csv")
```

```python
pd.set_option("display.max_columns",None)
```

```python
df.head()
```

```python
# Spliting data into independent variable and dependent variable 
X = df.drop("price",axis = 1)  #  independent variable
y = df["price"] #  dependent variable
```

```python
print("Shape of dataframe : ",df.shape)
print("Shape of X : ",X.shape)
print("Shape of y : ",y.shape)
```

```python
# train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, random_state = 5,test_size = 0.2)
```

```python
print("Shape of X_train : ",X_train.shape)
print("Shape of y_train : ",y_train.shape)
print("Shape of X_test : ",X_test.shape)
print("Shape of y_test : ",y_test.shape)
```

```python
# Feature Scaling of data 
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# need to give degree
poly_reg = PolynomialFeatures(degree=2)  
# degree = 2 will create feature of every degree upto 2 such that 0,1,2 degree

# example
# X [[2,3]]  , here x=2 and y=3   are 2 features
# [1,x,y,x^2,xy,y^2]   => elaburate => [1,2,3,4,6,9]


# degree = 2 then features should we x1 and x1^2 
poly_reg.fit(X_train)
X_train_poly = poly_reg.transform(X_train)
X_test_poly = poly_reg.transform(X_test)
```

```python
print(X_train_poly.shape)  # 2-D array
print(X_test_poly.shape)    # 2-D array
# NOw data is transformed into 5886 features initially their was only 107 features 
# NOw we have to use Linear Regression model to train data
```

```python
# Now use Linear Regression model

lr = LinearRegression()
lr.fit(X_train_poly,y_train)  # it will take some time  to train model bcz of size of data
```

```python
lr.score(X_test_poly,y_test)  # near about score is 1 this score is best 
```

```python
# predict
lr.predict([X_test_poly[0,:]])
```

```python
y_pred = lr.predict(X_test_poly)
y_pred
```

```python
y_test
```

```python
# the model is giving me proper prediction
```

```python
# To find error.
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
```

```python
print("MSE : ",mse)
print("RMSE : ",rmse)
```
