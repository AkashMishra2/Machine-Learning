```python
# ID3 Entropy is use to find the impurity/uncertainity in the dataset.

# With mathematical feature for information gain...
# formula is :
#             Entropy = (-p1 log2(p1)-p2 log2(p2))    for 2 output calsses like "YES"/"NO".
# for mulitple classes
#             Entropy  = -summation(pn log2(pn))    .... from i=1 to n number of output classes 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading dataset 
dataset = pd.read_csv(r"D:\Social_Network_Ads.csv")
dataset.head(10)
```

```python
print("shape of dataset:",dataset.shape)
```

```python
# Prepare the data through preprocessing.

# Step:- 1

# Split the data in to parts X,y

# input 
X = dataset.iloc[:,2:4]
# output
y = dataset["Purchased"]
```

```python
# Step:- 2
# Train and test 

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Step:- 3
# Feature scale the dataset

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = sc.transform(X_test)
```

```python
X_test
```

```python
# Creating the model 
#  Here we will use Entropy.

from sklearn.tree import DecisionTreeClassifier
```

```python
dtid3 = DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
dtid3.fit(X_train,y_train)
```

```python
y_pred = dtid3.predict(X_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
# find the accuracy : 
# But accuracy is not always the best.
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Accuracy : ",accuracy_score(y_test,y_pred))
```

```python
cf = confusion_matrix(y_test,y_pred)
```
