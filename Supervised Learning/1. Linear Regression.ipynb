```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression  # one class for simple and multiple linear regression
```

```python
# Loading data 
df = pd.read_csv(r"D:\bangalore house price prediction OHE-data.csv")
df
```

```python
# total_sqft	bath	balcony	price
X = df.drop(["price"],axis = 1)
y = df["price"]

print(X.shape)
print(y.shape)
```

```python
# spliting data for train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# when you see the data carefully you will see that the range of value in features are different 
# so we have to do feature scaling

sc = StandardScaler()

sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Train the Machine Learning 
lr = LinearRegression()

# train 
lr.fit(X_train,y_train)  # it finds the parameters of Linear Regression
```

```python
# now model has learned
print(lr.coef_) # cooeficient
print("\n")
print(lr.intercept_)  # intercept
```

```python
# train has done 
# But Now I am give information about new house 
# I want to pridict the price of that new house

X_test[0, :]  # this is the information of single house
```

```python
# above we have given the information about a house 
# and want to predict the price of house  using 

#   <object>.predict([<information of new house>])  parameter should be 2-Dimensional data 

lr.predict([X_test[0,:]])
```

```python
print(lr.predict(X_test))
```

```python
# to check the accuracy of Ml Model 
# <object>.score(X_test<predicted value>,y_test<actual value>)

lr.score(X_test,y_test)

# regression model giving us value with 79.038 % of accuracy
```
